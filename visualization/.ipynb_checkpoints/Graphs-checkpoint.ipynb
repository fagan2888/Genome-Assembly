```python
files = ["881_genomecov.tsv", "882_genomecov.tsv",
         "887_genomecov.tsv", "888_genomecov.tsv",
         "889_genomecov.tsv", "890_genomecov.tsv"]
genes = ['arnA', 'arnB', 'arnC', 'arnD', 'arnE', 'arnF', 'arnT', 'eptA', 'phoP', 'PhoQ', 'pmrA', 'PmrB']

# Make a dictionary of pandas DataFrames with filenames as keys
df_dict = {}
for f in files:
    df = pd.read_csv(f, sep='\t', names=['gene', 'position', 'coverage'] ) #names are for the referencing
    df_dict[f] = df
```

```python
for gene in genes:
    fig, ax = plt.subplots(1, figsize=(10,7))
    ax.set_title(gene)
    ax.set_xlabel('position')
    ax.set_ylabel('coverage')
    for file in files:
        df = df_dict[file]
        gene_df = df[df['gene'].str.contains(gene)]
        pos, cov = gene_df['position'], gene_df['coverage']
        
        ax.plot(pos, cov, label=file)
```

```python
for gene in genes:
    mean_cov = []
    fileIDs = []
    for file in files:
        df = df_dict[file]
        gene_df = df[df['gene'].str.contains(gene)]
        cov = gene_df['coverage'].mean()
        mean_cov.append(cov)
        
        fileID = file[0:3]
        fileIDs.append(fileID)
    mean_df = pd.DataFrame(data=mean_cov, index=fileIDs)
    
    mean_df.plot(kind='bar', legend=False, title=gene)
```
